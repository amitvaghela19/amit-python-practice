```python
#change (whitspace,not avalable, specific words are used in some files  for null value-for example "Eat Melody") to NaN. If you want to write anything else than Nan, use replace.

import pandas as pd
df= pd.read_csv('new_weather_data.csv',na_values=("notavaialble","Eat Melody"))
df=df.replace(["disha"],"Direction")
df
```
