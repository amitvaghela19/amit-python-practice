```python
## Initially created if...elif statement but
try:
    temperature_input = input("Enter temperature in celsius: ")
    temperature = int(temperature_input)
    print("Temperature is", temperature, "Celsius")

    if 10 < temperature < 35:
        print("Today is a good day")

    elif 0 <= temperature < 10:
        print("Today is a cool day")

    elif -10 <= temperature < 0:
        print("Stay warm")

    elif -20 <= temperature < -10:
        print("Yellow Warning: Stay Home")

    elif -40 <= temperature < -20:
        print("Red Alert: Please keep your mobile charged all the time. "
              "Keep the water running. Call 911 in case of emergency, "
              "service may delay depending on the condition")

    elif 35 <= temperature < 50:
        print("Too Hot")

    elif temperature > 50 or temperature < -40:
        print("Are You An Alien?")

except ValueError:
    print("Invalid...............Please enter only numbers, for example: 10, 18, -1, 0, -18 etc...")
```
